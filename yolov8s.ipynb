```python
import os
import h5py
import cv2
import numpy as np
from tqdm import tqdm

DATASET_ROOT = "/content/s1s2_landslide_reference_data/reference_data"
OUTPUT_DIR = "/content/yolo_dataset"

os.makedirs(f"{OUTPUT_DIR}/images/train", exist_ok=True)
os.makedirs(f"{OUTPUT_DIR}/images/val", exist_ok=True)
os.makedirs(f"{OUTPUT_DIR}/labels/train", exist_ok=True)
os.makedirs(f"{OUTPUT_DIR}/labels/val", exist_ok=True)

def process_file(h5_path, split):

    with h5py.File(h5_path, "r") as f:

        red = f["POST1_B04"][:]
        green = f["POST1_B03"][:]
        blue = f["POST1_B02"][:]
        mask = f["None_MASK"][:]

        total = red.shape[0]

        for i in tqdm(range(total)):

            r = red[i,0]
            g = green[i,0]
            b = blue[i,0]

            img = np.stack([r,g,b], axis=-1)

            img = cv2.normalize(img,None,0,255,cv2.NORM_MINMAX).astype(np.uint8)

            m = mask[i,0]

            contours,_ = cv2.findContours(
                (m>0).astype(np.uint8),
                cv2.RETR_EXTERNAL,
                cv2.CHAIN_APPROX_SIMPLE
            )

            h,w = img.shape[:2]

            labels = []

            for c in contours:

                x,y,ww,hh = cv2.boundingRect(c)

                xc = (x + ww/2)/w
                yc = (y + hh/2)/h
                ww /= w
                hh /= h

                labels.append(f"0 {xc} {yc} {ww} {hh}")

            name = f"{split}_{i}"

            cv2.imwrite(f"{OUTPUT_DIR}/images/{split}/{name}.png", img)

            with open(f"{OUTPUT_DIR}/labels/{split}/{name}.txt","w") as ftxt:
                ftxt.write("\n".join(labels))


process_file(f"{DATASET_ROOT}/train_n3_s1s2.h5","train")
process_file(f"{DATASET_ROOT}/val_n3_s1s2.h5","val")

print("Dataset conversion finished.")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")   # load pretrained model

model.train(
    data="yolo_dataset/dataset.yaml",
    epochs=150,
    imgsz=256
)
```
